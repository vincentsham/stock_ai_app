```python
# Query the `stock_metadata` table
conn = connect_to_db()
# show all columns
pd.set_option('display.max_columns', None)
if conn:
    query = \
    """
    WITH latest_earnings AS (
        SELECT 
            tic, calendar_year, calendar_quarter, 
            eps, eps_estimated, revenue, revenue_estimated 
        FROM core.earnings 
        WHERE tic = 'TSLA' AND eps IS NOT NULL 
        ORDER BY tic, calendar_year DESC, calendar_quarter DESC 
    LIMIT 1)
    SELECT eps_surprise_regime, revenue_surprise_regime
    FROM core.earnings_metrics em
    JOIN latest_earnings le
    ON em.tic = le.tic
    AND em.calendar_year = le.calendar_year
    AND em.calendar_quarter = le.calendar_quarter
    WHERE le.tic = 'TSLA'
    ;
    """
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    # Display the data
    display(df)
```

```python
# Query the `stock_metadata` table
conn = connect_to_db()
# show all columns
pd.set_option('display.max_columns', None)
# show all rows
pd.set_option('display.max_rows', None)
if conn:
    query = \
    """
    WITH latest_date AS (
    SELECT date
    FROM raw.stock_ohlcv_daily
    WHERE tic = 'TSLA'
    ORDER BY date DESC
    LIMIT 1)
    SELECT tic, end_date as date, pt_high, pt_low, pt_p25, pt_median, pt_p75
    FROM core.analyst_rating_quarterly_summary AS arqs
    JOIN latest_date ld
    ON arqs.end_date >= ld.date - INTERVAL '1 year' AND arqs.end_date <= ld.date
    WHERE arqs.tic = 'TSLA'
    ORDER BY date DESC;
    """
    _df = pd.read_sql_query(query, conn)
    conn.close()
    
    # Display the data
    
    display(_df)
```

```python
# Query the `stock_metadata` table
conn = connect_to_db()
# show all columns
pd.set_option('display.max_columns', None)
if conn:
    query = \
    """
    WITH latest_date AS (
        SELECT date
        FROM raw.stock_ohlcv_daily
        WHERE tic = 'TSLA'
        ORDER BY date DESC
        LIMIT 1)
    SELECT sod.tic, sod.date, sod.close
    FROM raw.stock_ohlcv_daily sod
    JOIN latest_date ld
    ON sod.date >= ld.date - INTERVAL '1 year' AND sod.date <= ld.date
    WHERE sod.tic = 'TSLA'
    ORDER BY sod.date DESC
    ;
    """
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    # Display the data
    display(df)
```
