```python
import duckdb

# URL to the file
url = "https://huggingface.co/datasets/bwzheng2010/yahoo-finance-data/resolve/main/data/stock_historical_eps.parquet"

print("Querying remote Parquet file via DuckDB...")

# Create an in-memory database connection
con = duckdb.connect()

# Install and load the httpfs extension to access remote URLs (happens automatically in newer versions)
con.execute("INSTALL httpfs; LOAD httpfs;")

# --- QUERY 1: Get all data for a specific stock (e.g., NVDA) ---
# Note: We query the URL directly as if it were a table
query = f"SELECT * FROM '{url}' WHERE symbol = 'NVDA' ORDER BY quarter_date DESC"

# Execute and fetch as a Pandas DataFrame
df = con.execute(query).df()

print(f"\n--- Results for NVDA ({len(df)} rows) ---")
print(df.head())

# --- QUERY 2: Complex Analysis (e.g., Average Surprise % by Quarter) ---
# DuckDB is extremely fast for aggregations like this
# agg_query = f"""
#     SELECT 
#         quarter_name,
#         AVG(CAST(replace(surprise_percent, '%', '') AS FLOAT)) as avg_surprise
#     FROM '{url}'
#     WHERE symbol IN ('AAPL', 'MSFT', 'GOOG')
#     GROUP BY quarter_name
#     ORDER BY quarter_name DESC
#     LIMIT 5
# """

# print("\n--- Aggregation Example ---")
# print(con.execute(agg_query).df())
```
