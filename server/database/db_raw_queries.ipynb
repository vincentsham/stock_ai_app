```python
conn = connect_to_db()
if conn:
    query = "SELECT * FROM raw.analyst_price_targets;"
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    # Display the data
    display(df)
```

```python
import numpy as np

# df columns expected:
# ['tic','published_at','title','site','analyst_name','company',
#  'price_target','adj_price_target','price_when_posted','url',...]
# Make sure published_at is datetime
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
df = df.sort_values(['tic', 'analyst_name', 'published_at'])

# Choose adjusted PT when available
df['pt'] = df['adj_price_target'].fillna(df['price_target'])

# Sort so shift() is deterministic (tie-break with url)
df = df.sort_values(['tic','analyst_name','published_at','url'], kind='mergesort')

# Previous PT within (tic, analyst); include 'company' too if you want stricter scoping
grp_cols = ['tic','analyst_name']
df['prev_pt'] = df.groupby(grp_cols, sort=False)['pt'].shift(1)
df['prev_published_at'] = df.groupby(grp_cols, sort=False)['published_at'].shift(1)
df['prev_url'] = df.groupby(grp_cols, sort=False)['url'].shift(1)

df['pt_delta'] = df['pt'] - df['prev_pt']
df['pt_change'] = np.select(
    [
        df['prev_pt'].isna(),
        df['pt'] > df['prev_pt'],
        df['pt'] < df['prev_pt'],
    ],
    ['first_seen','upgrade','downgrade'],
    default='reiterate'
)

per_entry_changes = df[
    ['tic','analyst_name','company','published_at','url',
     'prev_published_at','prev_url','prev_pt','pt','pt_delta','pt_change']
].copy()

df = df.sort_values(['tic','published_at'])
```
